## Theory of Signal Operations

Signal operations are fundamental manipulations applied to signals in various domains like signal processing, communications, and control systems. These operations allow us to modify, analyze, and synthesize signals for different purposes. Here are some common signal operations:

### 1. Amplitude Scaling

Amplitude scaling involves multiplying the amplitude of a signal by a constant factor. If $x(t)$ is a continuous-time signal and $A$ is the scaling factor, the scaled signal $y(t)$ is given by:

$y(t) = A \cdot x(t)$

For a discrete-time signal $x[n]$, the scaled signal $y[n]$ is:

$y[n] = A \cdot x[n]$

This operation changes the strength or intensity of the signal.

### 2. Time Shifting

Time shifting involves moving the signal along the time axis. A positive shift moves the signal to the right (delay), while a negative shift moves it to the left (advance).

For a continuous-time signal $x(t)$, a time-shifted signal $y(t)$ is:

$y(t) = x(t - t_0)$

where $t_0$ is the amount of time shift. If $t_0 > 0$, it's a delay; if $t_0 < 0$, it's an advance.

For a discrete-time signal $x[n]$, a time-shifted signal $y[n]$ is:

$y[n] = x[n - n_0]$

where $n_0$ is the number of samples shifted.

### 3. Time Reversal

Time reversal involves flipping the signal about the vertical axis (time origin). If $x(t)$ is a continuous-time signal, the time-reversed signal $y(t)$ is:

$y(t) = x(-t)$

For a discrete-time signal $x[n]$, the time-reversed signal $y[n]$ is:

$y[n] = x[-n]$

This operation is useful for analyzing the symmetry properties of a signal.

### 4. Time Scaling

Time scaling involves compressing or expanding the signal along the time axis.

For a continuous-time signal $x(t)$, a time-scaled signal $y(t)$ is:

$y(t) = x(at)$

where $a$ is the scaling factor. If $a > 1$, the signal is compressed; if $0 < a < 1$, the signal is expanded. If $a < 0$, it involves both time scaling and time reversal.

For a discrete-time signal $x[n]$, time scaling is a bit more complex and often involves interpolation or decimation. A simple form of time scaling for discrete-time signals is:

$y[n] = x[kn]$

where $k$ is an integer. If $k > 1$, it's decimation (sampling the signal at a lower rate); if $0 < k < 1$ (and $1/k$ is an integer), it's interpolation (inserting samples).

### 5. Addition

Addition of two signals involves adding their amplitudes at each point in time. If $x_1(t)$ and $x_2(t)$ are two continuous-time signals, their sum $y(t)$ is:

$y(t) = x_1(t) + x_2(t)$

For discrete-time signals $x_1[n]$ and $x_2[n]$, their sum $y[n]$ is:

$y[n] = x_1[n] + x_2[n]$

Signals must have the same time base for meaningful addition.

### 6. Multiplication

Multiplication of two signals involves multiplying their amplitudes at each point in time. If $x_1(t)$ and $x_2(t)$ are two continuous-time signals, their product $y(t)$ is:

$y(t) = x_1(t) \cdot x_2(t)$

For discrete-time signals $x_1[n]$ and $x_2[n]$, their product $y[n]$ is:

$y[n] = x_1[n] \cdot x_2[n]$

This operation is used in modulation, mixing, and other signal processing applications.

These are some of the basic signal operations. More complex operations can be built by combining these fundamental ones.

4. Write code to generate and sketch following sequence
g[n] = 5,n>=0
g[n] = 5-3n,0<n<=4
g[n] = -23+n^2, 4<n<=8
g[n] = 41,n>8


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of n
n = np.arange(0, 15)  # Covers n from 0 to 14

# Initialize g[n] with zeros
g = np.zeros_like(n, dtype=float)

# Apply piecewise conditions
for i in range(len(n)):
    if n[i] == 0:
        g[i] = 5
    elif 0 < n[i] <= 4:
        g[i] = 5 - 3 * n[i]
    elif 4 < n[i] <= 8:
        g[i] = -23 + n[i]**2
    elif n[i] > 8:
        g[i] = 41

# Plotting the sequence
plt.figure(figsize=(10, 5))
plt.stem(n, g, basefmt=" ")
plt.title('Sushant kapil, 9923102039 - Discrete Sequence g[n]')
plt.xlabel('n')
plt.ylabel('g[n]')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original signal function
def x(t):
    return t * np.exp(-t)

# Time vector
t = np.linspace(-1, 1, 400)

# Signal transformations
signals = {
    'a. Sushant kapil , 9923102039 - x(t)': x(t),
    'b. Sushant kapil , 9923102039 - x(-t)': (-t) * np.exp(t),
    'c. Sushant kapil , 9923102039 - x(t/5)': x(t / 5),
    'd. Sushant kapil , 9923102039 - x(5t)': x(5 * t),
    'e. Sushant kapil , 9923102039 - x(t-3)': x(t - 3),
    'f. Sushant kapil , 9923102039 - x(1-2t)': x(1 - 2 * t)
}


# Plotting
plt.figure(figsize=(12, 10))
for i, (title, y) in enumerate(signals.items(), 1):
    plt.subplot(3, 2, i)
    plt.plot(t, y)
    plt.title(title)
    plt.grid(True)
    plt.xlabel('t')
    plt.ylabel('Amplitude')

plt.tight_layout()
plt.suptitle('Sushant kapil , 9923102039 - Signal Transformations of x(t) = t * exp(-t)', fontsize=16, y=1.02)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def u(t):
    return np.where(t >= 0, 1, 0)

def v(t):
    # Piecewise construction using unit step (Heaviside) functions
    return (10 * u(t) # from t = 0
            - 10 * u(t - 2) # step down at t = 2
            + 20 * u(t - 2) # step up at t = 2
            - 20 * u(t - 3) # step down at t = 3
            - 10 * (t - 3) * (u(t - 3) - u(t - 5)) # linear decrease from t = 3 to t = 5
            + 10 * (t - 5) * (u(t - 5) - u(t - 7))) # linear increase from t = 5 to t = 7

t = np.linspace(0, 7, 700)
V = v(t)

plt.plot(t, V)
plt.xlabel('t (s)')
plt.ylabel('v(t) (V)')
plt.title('Voltage waveform v(t) as sum of unit step functions - SUSHANT KAPIL, 9923102039')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original signal v(t): a rectangular pulse from -1 to 1
def v(t):
    return np.where((t >= -1) & (t <= 1), 1, 0)

# Time axis
t = np.linspace(-5, 5, 1000)

# Transformations
v_neg_t = v(-t)         # a. Time reversal
v_scaled_shifted = 3*v(t + 1)  # b. Amplitude scaling and time shift
v_compressed = v(4*t)   # c. Time compression
v_shifted_scaled = -2*v(t - 1) # d. Time shift and amplitude scaling

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(t, v(t), label='v(t)')
plt.title('Original Signal v(t) - sushant kapil ,9923102039')
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 2)
plt.plot(t, v_neg_t, label='v(-t)', color='orange')
plt.title('a. Time Reversal - sushant kapil ,9923102039')
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 3)
plt.plot(t, v_scaled_shifted, label='3v(t+1)', color='green')
plt.title('b. Amplitude & Shift - sushant kapil ,9923102039')
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 4)
plt.plot(t, v_compressed, label='v(4t)', color='red')
plt.title('c. Time Compression - sushant kapil ,9923102039')
plt.grid(True)
plt.legend()

plt.subplot(2, 3, 5)
plt.plot(t, v_shifted_scaled, label='-2v(t-1)', color='purple')
plt.title('d. Shift & Scaling - sushant kapil ,9923102039')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
